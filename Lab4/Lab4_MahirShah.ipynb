### Q1

a) Load data found in Lab4.xlsx into dataframe using pandas. Print the first 10 records of each tab.

```python
#your solution
import pandas as pd

assignment = pd.read_excel('Lab4.xlsx', sheet_name='Assignment')
graded = pd.read_excel('Lab4.xlsx', sheet_name='Graded')
test = pd.read_excel('Lab4.xlsx', sheet_name='Test')
exam = pd.read_excel('Lab4.xlsx', sheet_name='Exam')
print(assignment[:11])
print(graded[:11])
print(test[:11])
print(exam[:11])
```

b) Find the mean score of each assessment component per student. Store the Student ID, Mean Assignment score, Mean Graded Score, Mean Test Score and Mean Exam score in one dataframe.
Print the dataframe.

```python
#your solution
df = pd.concat([assignment[1:].mean(axis=1), graded[1:].mean(axis=1), test[1:].mean(axis=1), exam[1:].mean(axis=1)], axis=1)
df.index = assignment.iloc[1:,0]
df.columns = ['Assignment', 'Graded', 'Test', 'Exam']
df
```

```python
#your solution
df = pd.DataFrame(pd.concat([assignment[1:].mean(axis=0), graded[1:].mean(axis=0), test[1:].mean(axis=0), exam[1:].mean(axis=0)], axis=0), columns=['Mean Score'])
df
```

Web scraping is the automated collection of data from websites (to be more precise, from the HTML content of websites).

The commands below will give us a nested data structure of the original HTML content from the website, stored in variable `soup`, which we can navigate and collect data from.

```python
from bs4 import BeautifulSoup as bs
import requests
url = "https://www.bookdepository.com/bestsellers"
response = requests.get(url)
html = response.content
soup = bs(html, "lxml")
soup
```
