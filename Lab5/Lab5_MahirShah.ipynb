### Q1

a) Load data found in Lab5.xlsx into dataframe using pandas. Combine the data in both tabs into one dataframe. The final merged table should be a union of data from both tabs.

```python
#your solution
import pandas as pd
import numpy as np
df = pd.read_excel('Lab5.xlsx', sheet_name='Normal').merge(pd.read_excel('Lab5.xlsx', sheet_name='Overtime'), on='name', how='outer')
df
```

b) Summarize the number of missing value for each column.

```python
#your solution
df.isna().sum()
```

c) Drop the rows that has missing value in the 'Staff ID' column.

```python
#your solution
df.dropna(subset=["Staff ID"], axis=0, inplace= True) 
df
```

d) Replace all the blanks in any of the columns with 'NaN'.

```python
#your solution
df = df.replace(np.nan, 'NaN')
df = df.replace('-', 'NaN')
df
```

e) Remove all rows with 'Nan' in column over_time_work.

```python
#your solution
df = df[~(df.over_time_work == 'NaN')]
df
```

f) Reset the index of the dataframe.

```python
#your solution
df = df.reset_index().drop(columns='index')
df
```

g) Rename the column over_time_work to overtime_hrs_work

```python
#your solution
df = df.rename(columns={'over_time_work' : 'overtime_hrs_work'})
df
```

h) Change the data type of StaffID to int.

```python
#your solution
df['Staff ID'] = df['Staff ID'].astype(int)
df
```

i) Catagorize the data in `normal_hrs_work` into 3 bins (High, Moderate, Low) using equal width binning. Add a column named `normal_hrs_work_bin` with the respective labels (High, Moderate, Low).

```python
#your solution
df['normal_hrs_work_bin'] = pd.cut(df[~(df.normal_hrs_work == 'NaN')]['normal_hrs_work'], 3, labels=['Low', 'Moderate', 'High'], include_lowest=True)
df
```
